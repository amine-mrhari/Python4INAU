## 1. Variables

### 1.1 Définition

Une variable est un espace mémoire nommé permettant de stocker une valeur. Elle sert à conserver des données que l'on pourra réutiliser ou modifier tout au long du programme.

### 1.2 Syntaxe

Pour déclarer une variable en Python, on utilise le signe égal (`=`) pour assigner une valeur à un nom de variable.


### 1.3 Règles de Nommage

- Le nom doit commencer par une lettre ou un underscore (`_`).
- Pas d’espaces, on utilise `snake_case` (mots en minuscules séparés par `_`).
- La casse est significative (Python est sensible à la casse).
- Noms explicites, pas de noms comme `a`, `x`.
- Ne pas commencer par un chiffre.
- Éviter les mots réservés (comme `def`, `class`, `import`, etc.).

## 2. Types de Données

### 2.1 Types de Base

Python supporte plusieurs types de données de base :

- **Entiers (int)** : Nombres entiers, par exemple `42`, `-7`.
- **Flottants (float)** : Nombres à virgule flottante, par exemple `3.14`, `-0.001`.
- **Chaînes de caractères (str)** : Texte, par exemple `"Bonjour"`, `"Python"`.
- **Booléens (bool)** : Valeurs logiques, `True` ou `False`.

### 2.2 Exemples

```python
# Exemples de types de données
entier = 10
flottant = 3.14
chaine = "Hello, World!"
booleen = True
```

### 2.3 Vérification du Type

Vous pouvez vérifier le type d'une variable en utilisant la fonction `type()`.

```python
print(entier, type(entier))       # <class 'int'>
print(flottant, type(flottant))   # <class 'float'>
print(chaine, type(chaine))       # <class 'str'>
print(booleen, type(booleen))     # <class 'bool'>
```

## 3. Opérations

### 3.1 Opérations Arithmétiques

Python permet d'effectuer diverses opérations arithmétiques sur les nombres :

- Addition : `+`
- Soustraction : `-`
- Multiplication : `*`
- Division : `/`
- Division entière : `//`
- Modulo : `%`
- Puissance : `**`

### 3.2 Exemples d'Opérations

```python
# Exemples d'opérations arithmétiques
a = 10
b = 3

addition = a + b        # 13
soustraction = a - b    # 7
multiplication = a * b  # 30
division = a / b        # 3.333...
division_entiere = a // b # 3
modulo = a % b          # 1
puissance = a ** b      # 1000
```

### 3.3 Opérateurs d’assignation combinée (Compound Assignment Operators)
Ces opérateurs permettent d’effectuer une opération mathématique tout en réaffectant la variable.

```python
# Exemple de base
valeur = 200

# Ajouter 100 à la valeur existante
valeur += 100
print(valeur)  # Résultat : 300

# Soustraire 100
valeur -= 100
print(valeur)  # Résultat : 200

# Multiplier par 20 (par exemple, 200 unités foncières par 20 habitants)
valeur *= 20
print(valeur)  # Résultat : 4000

# Diviser par 2 (par exemple, répartir la population sur 2 zones)
valeur /= 2
print(valeur)  # Résultat : 2000.0
```

### 6 Opérations logiques et de comparaison

### 6.1 Type de données booléen (Boolean)

Le type bool en Python représente une valeur logique : True (vrai) ou False (faux).

```python
print(bool(1))         # True → 1 est considéré comme vrai
print(bool(0))         # False → 0 est considéré comme faux
print(bool(True))      # True → c’est déjà une valeur booléenne vraie
print(bool(False))     # False → c’est déjà une valeur booléenne fausse
```

### 6.2 Les opérateurs de comparaison en Python

Les opérateurs de comparaison permettent de comparer deux valeurs (nombres, chaînes, etc.). Ils retournent toujours un booléen : True ou False.

```python
a = 100
b = 200

print(a == b)   # False → 100 n'est pas égal à 200
print(a != b)   # True → 100 est différent de 200
print(a > b)    # False → 100 n'est pas supérieur à 200
print(a < b)    # True → 100 est inférieur à 200
print(a >= b)   # False → 100 n'est pas supérieur ou égal à 200
print(a <= b)   # True → 100 est inférieur ou égal à 200
```

Autres exemples :

```python
print(bool(-1))        # True → tout nombre différent de 0 est vrai
print(bool(0))         # False
print(bool(None))      # False → None signifie "aucune valeur"
print(bool(''))        # False → une chaîne vide est fausse
print(bool('hello'))   # True → une chaîne non vide est vraie
print(bool([]))        # False → une liste vide est fausse
print(bool([10, 20]))  # True → une liste avec des éléments est vraie
```

### 6.3 Chaînes de caractères et opérateurs de comparaison

```python
print('aaa' == 'aaa')      # True → les deux chaînes sont identiques
print('aaa' == 'bbb')      # False → les chaînes sont différentes
print('aaa' != 'aaa')      # False → elles sont identiques donc pas différentes
print('aaa' != 'bbb')      # True → elles sont bien différentes
```

Les chaînes de caractères peuvent être comparées avec == (égal) et != (différent).

**Remarque importante**
Les chaînes de caractères peuvent aussi être comparées, mais attention à l’ordre alphabétique ASC

```python
print("Résultat de 'a' > 'b' :", 'a' > 'b')  # False
```

### 6.4 Différence entre == et is

```python
a = 1
b = 1.0

print(a == b)   # True → 1 est égal à 1.0 (même valeur)
print(a is b)   # False → mais pas le même type (entier vs flottant)
```

**==** compare les valeurs, tandis que **is** compare les identités en mémoire (donc aussi les types). Important à connaître mais rarement utilisé pour les valeurs simples.

### 6.5 L’opérateur (in)

Permet de vérifier si un élément est contenu dans un autre (souvent une chaîne ou une liste).

```python
print('aaa' in 'aaa-bbb-ccc')   # True → "aaa" est bien présent
print('bbb' in 'aaa-bbb-ccc')   # True → idem
print('ddd' in 'aaa-bbb-ccc')   # False → "ddd" n’apparaît pas
```

### 6.6 Les opérateurs : and, or, not

```python
x = True
y = False

print(x and y)    # False → les deux doivent être True
print(x or y)     # True  → au moins un est True
print(not x)      # False → inverse de True
print(not y)      # True  → inverse de False
```
