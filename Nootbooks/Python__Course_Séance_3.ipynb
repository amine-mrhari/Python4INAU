### 1. Instructions Conditionnelles : Conditions et Prise de Décision


### 1.1. Structure conditionnelle de base : **if**

En Python, l'instruction ```if``` permet d'exécuter un bloc de code uniquement si une condition est vraie. La syntaxe est la suivante :​

```python
     if condition:
       #Bloc d'instructions exécuté si la condition est vraie
```

if.jpeg

**Remarque :**

En Python, l'indentation (généralement **4 espaces**) est essentielle pour définir les blocs de code. Un mauvais alignement peut entraîner des erreurs de syntaxe.

**Exemple 1 :**

Ce programme vérifie si la densité de population d'une zone dépasse un seuil critique :

````python
# Demander à l'utilisateur de saisir la superficie en hectares
superficie = float(input("Entrez la superficie de la zone en hectares : "))

# Demander à l'utilisateur de saisir la population
population = int(input("Entrez la population de la zone : "))

# Calculer la densité de population (habitants par hectare)
densite = population / superficie

# Vérifier si la densité dépasse le seuil critique
if densite > 150:
    print("Attention : la densité dépasse le seuil critique de 150 hab/ha.")
print("************* Fin **************** ")
````

**Exemple 2 :** Analysons un quartier avec l'instruction **if**

On souhaite classifier une zone en fonction de sa densité de population (en habitants par km²).

- Afficher « Zone très dense » si la densité est strictement supérieure au seuil.

- Afficher « Zone peu dense » sinon (c’est-à-dire lorsque la densité est inférieure ou égale au seuil).

````python
densite = 18500  # habitants par km²
seuil_dense = 10000

if densite > seuil_dense:
    print("Zone très dense")
if densite <= seuil_dense:
    print("Zone peu dense")
````

### 1.2 Structure conditionnelle de base : **if** et  **else**

Si la condition est fausse, le bloc d'instructions est ```ignoré```. Pour gérer plusieurs cas, on peut utiliser ```else``` (sinon) :

```python
    if condition:
      # Bloc exécuté si condition1 est vraie
    else:
      # Bloc exécuté si aucune des conditions précédentes n'est vraie

 ### 1.3 Structure conditionnelle de base : **if**, **elif**, **else**

 Pour gérer plusieurs cas, on peut utiliser ```elif``` (sinon si) et ```else``` (sinon) :
```python
    if condition1:
      # Bloc exécuté si condition1 est vraie
    elif condition2:

    else:

**Exemple 3 :** Analysons un quartier de Fes avec  les instructions **if** et **else**

````python
densite = 5001  # habitants par km²
seuil_dense = 20000

if densite > seuil_dense:
    print("Zone très dense")
elif densite > 8000:
    print("Densité moyenne")
elif densite > 5000:
    print("Densité moins ")
else:
    print("Zone peu dense")
````

### 1.4 Instructions conditionnelles imbriquées

Les instructions conditionnelles peuvent être ```imbriquées``` les unes dans les autres pour gérer des situations plus complexes.​

**Exemple :** le programme évalue d'abord si la densité est supérieure à 5000 habitants par km². Si c'est le cas, il vérifie ensuite la superficie pour déterminer la catégorie de la zone.

````python
densite = 8500  # habitants par km²
superficie = 6  # en km²

if densite > 5000:
    if superficie < 1:
        print("Zone très dense et petite superficie")
    elif superficie < 5:
        print("Zone dense avec superficie modérée")
    else:
        print("Zone dense et grande superficie")
else:
    print("Zone peu dense")
````

**Exemple 1 :** Jeu de simulation :

On souhaite simuler un jeu de lancer de pièce pour déterminer aléatoirement si l’on se trouve dans une zone résidentielle ou commerciale.

## 3. Exercice : Classification d’un quartier

### Énoncé  
Vous devez écrire un programme qui génère aléatoirement la population et la superficie d’un quartier, puis détermine son type selon la densité et signale l’absence éventuelle de services de base

### Affichage :

    Population : 264568 hab.
    Superficie : 18.98 km²
    Densité    : 13939.30 hab./km²
    Type       : urbain
    École      : Présent
    Hôpital    : Manquant
    Mosquée    : Manquant


### Consignes

Commencez par importer `randint` et `uniform` depuis le module `random`. Générez un entier `population` compris entre 1 000 et 500 000. Générez un nombre à virgule `superficie` compris entre 0,5 et 100, arrondi à deux décimales.  

Ensuite, créez trois variables booléennes `a_ecole`, `a_hopital` et `a_mosquee` en utilisant `random.choice([True, False])` afin de simuler la présence ou l’absence de chacun de ces services.  

Calculez la densité en divisant la population par la superficie. Si `superficie` était nulle, votre programme devra capturer l’erreur et afficher « Erreur : superficie nulle ». Si un des types n’est pas numérique, afficher « Erreur de type ».  

Pour déterminer le type de quartier, comparez la densité à deux seuils fixes : si elle est strictement inférieure à 100, considérez-le rural ; si elle est comprise entre 100 inclus et 500 exclus, considérez-le périurbain ; sinon, considérez-le urbain.  

Enfin, affichez la population générée, la superficie, la densité (avec deux décimales), le type de quartier, puis, pour chacun des trois services, un message indiquant « Présent » ou « Manquant ».  

````python
from random import randint, uniform, choice

# Génération des données
population = randint(1000, 500000)
superficie = round(uniform(0.5, 100), 2)

# Présence des services
a_ecole = choice(['Manquant', 'Présent'])
a_hopital = choice(['Manquant', 'Présent'])
a_mosquee = choice(['Manquant', 'Présent'])

# Calcul de densité
densite = population / superficie

# Détermination du type de quartier
if densite < 100:
    type_quartier = "rural"
elif densite < 500:
    type_quartier = "périurbain"
else:
    type_quartier = "urbain"

# Affichage des résultats

print(f"Population : {population} hab.")
print(f"Superficie : {superficie} km²")
print(f"Densité    : {densite:.2f} hab./km²")
print(f"Type       : {type_quartier}")
print(f"École      : {a_ecole}")
print(f"Hôpital    : {a_hopital}")
print(f"Mosquée    : {a_mosquee}")
````

### 4.2 Attribution aléatoire de tâches

Écrire un programme qui attribue aléatoirement des tâches (nettoyage, jardinage, surveillance) à des employés.



````python
# Attribution aléatoire unique d'une tâche différente pour chaque employé, sans boucle
import random

employes = ["Ahmed", "Fatima"]
taches = ["nettoyage", "jardinage", "surveillance"]

# S'assurer qu'on a le même nombre d'employés et de tâches
if len(employes) == len(taches):
    

    # Mélanger la liste des tâches
    random.shuffle(taches)
    
    
    # Affichage sans boucle
    print("Attribution aléatoire des tâches :")
    print(f"- {employes[0]} : {taches[0]}")
    print(f"- {employes[1]} : {taches[1]}")
    print(f"- {employes[2]} : {taches[2]}")
````

### 4.3 Jeu de devinette

Écrire un programme où l'utilisateur doit deviner un nombre entre 1 et 20 généré aléatoirement.

````python
import random

# Générer un nombre aléatoire entre 1 et 20
nombre_secret = random.randint(1, 20)

print("Je pense à un nombre entre 1 et 20.")

# Demander à l'utilisateur de proposer un nombre
essai = input("Devinez le nombre : ")

# Vérifier si l'utilisateur a entré un nombre entier
if essai.isdigit():
    essai = int(essai)
    if 0 < essai < 21 :
        
        if essai == nombre_secret:
            print("Bravo ! Vous avez deviné le bon nombre.")
        elif essai < nombre_secret:
            print("Trop petit. Le nombre était", nombre_secret)
        else:
            print("Trop grand. Le nombre était", nombre_secret)
    else :
        print("Entrée invalide. Veuillez entrer un nombre entre 1 et 20.")
else:
    print("Entrée invalide. Veuillez entrer un nombre entier.")
````
