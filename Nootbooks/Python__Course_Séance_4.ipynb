# 4.1 Boucle **for** et itération


### 4.1.1 Pourquoi itérer ?
Imaginons que vous deviez répéter une même action plusieurs fois, par exemple afficher un message d’analyse de densité urbaine. Sans boucle, vous écrivez :

````python
print("Analyse de la densité urbaine")
print("Analyse de la densité urbaine")
print("Analyse de la densité urbaine")
print("Analyse de la densité urbaine")
print("Analyse de la densité urbaine")
````

Avec une boucle **for**, on automatise cela très facilement :

python_for_loop.jpg

````python
for _ in range(5):
    print("Analyse de la densité urbaine")
````

### 4.1.2. Variable de contrôle (compteur)
On peut utiliser une variable pour suivre l’itération en cours. Cela est très utile pour numéroter ou indexer vos opérations.

````python
for i in range(5):
    print("– étude de la parcelle N°",i)
````

### 4.1.3  Quand anonymiser la variable ?
Si vous n’avez pas besoin du compteur, utilisez un underscore ```_``` pour indiquer que la variable est inutile :

````python
for _ in range(3):
    print("Génération d’un scénario aléatoire")
````

### 4.1.4 La fonction **range()**
La fonction ```range()``` génère une suite de nombres entiers. Sa syntaxe :
```
range(start=0, stop, step=1)

```

- ```start``` : valeur de départ (incluse)

- ```stop``` : valeur d’arrêt (exclue)

- ```step``` : pas d’incrément

### Exemples

````python
# Afficher les numéros de secteurs 1 à 5  (pas = 1)
for secteur in range(1, 6):
    print("Secteur", secteur)
````

````python
# Parcourir les parcelles de deux en deux (pas = 2)
for parcelle in range(2, 11, 2):
    print("Parcelle n°", parcelle)
````

````python
# Compte à rebours pour une simulation de démarrage
for t in range(5, 0, -1):
    print("Démarrage simulation dans", t, "secondes")
````

### 4.1.5 Somme cumulative : population totale de 10 secteurs

```
hypothétique population = numéro de secteur × 100
```

````python
pop_totale = 0
for i in range(1, 11):
    pop_totale += i * 100
print("Population totale des 10 secteurs :", pop_totale)
````

### 4.1.6 Itération sur un objet séquence (liste, chaîne…)
**Exemple avec une liste de quartiers :**

````python
quartiers = ["Médina", "Périphérie", "Centre-ville", "Zone industrielle"]
for q in quartiers:
    print("Étude du quartier :", q)
````

**Somme cumulée avec une liste de populations :**

````python
populations = [1200, 3400, 2750, 1800]
total = 0
for p in populations:
    total += p
print("Population totale :", total)
````

On peut aussi utiliser la fonction intégrée ```sum()``` :

````python
print("Population totale :", sum(populations))
````

**Itération sur une chaîne de caractères (par exemple un nom de projet) :**

````python
for lettre in "URBA2025":
    print(lettre, end=" ")
````

### 4.1.7 Boucles imbriquées (Exemple de combinaisons)
Imaginons qu’on veuille générer toutes les combinaisons possibles pour un projet : type de zone, mode de transport et service public associé :

````python
types_zone     = ["Résidentielle", "Commerciale"]
transports     = ["Métro", "Bus"]
services       = ["École", "Hôpital"]

print("Combinaisons de scénarios d’aménagement")
for z in types_zone:
    for t in transports:
        for s in services:
            print(f"{z} + {t} + {s}")
````

## 4.2 Introduction à la boucle **while**
### 4.2.1 Principe général
La boucle ```while``` exécute un bloc tant que la condition est vraie (condition à contrôle d’entrée).

- Si la condition est fausse dès le départ, le bloc n’est jamais exécuté.

- Utile pour des conditions variables, où on ne connaît pas à l’avance le nombre d’itérations.

### 4.2.2. Syntaxe
```
while condition:
    # bloc d’instructions répétées
```

**condition :** expression retournant True ou False

**indentation :** obligatoire pour délimiter le bloc

### 4.2.3. Schéma de fonctionnement

1.   Tester la condition
2.   Si ```True```, exécuter le bloc puis revenir à 1
3.   Si ```False```, sortir de la boucle

python_while_loop.jpg


### 4.2.4 Boucle infinie

Exemple de boucle infinie :

````python
i = 5
while i == 5:
    print("Boucle infinie")
````

**Danger :** si la condition reste toujours vraie, la boucle tourne sans fin.

**Comment l’éviter ?**

Toujours modifier une variable de contrôle dans la boucle, ou utiliser ```break```  pour interrompre.

### 4.3. Exemples concrets avec while
### 4.3.1. Comptage de parcelles à examiner
On veut examiner des parcelles numérotées 1 à 10.

````python
i = 1
while i <= 10:
    print(f"Analyse de la parcelle n°{i}")
    i += 1  # on incrémente i pour arrêter la boucle
````

### 4.3.2. Simulation de congestion routière
On simule un indice de congestion qui augmente chaque minute, et on s’arrête quand on dépasse 80 % (seuil critique).

````python
congestion = 0
minute = 0

while congestion < 80:
    minute += 1
    # simulation : la congestion croît de façon aléatoire
    congestion += 5  # ou random.randint(3, 10)
    print(f"Minute {minute} : congestion à {congestion}%")

print("Attention : seuil de congestion critique atteint.")
````

### 4.4. Utiliser **if** et **break** dans une boucle while
Pour interrompre la boucle avant que la condition principale ne devienne fausse, on peut placer un test interne avec break.

**Exemple :** arrêt anticipé en cas de détection d’une zone inondable

````python
zone = 1
while zone <= 20:
    print(f"Vérification de la zone {zone}")
    # simulons une alerte météo détectée sur la zone 7
    if zone == 7:
        print("Alerte inondation détectée ! Arrêt des vérifications.")
        break
    zone += 1
print("Boucle 'while' terminée.")
````

### 4.5. Bonnes pratiques et pièges
Toujours modifier la variable de contrôle dans la boucle pour éviter une boucle infinie.

Limiter l’imbrication des boucles pour garder le code lisible.

Utiliser ```else``` avec ```while``` pour exécuter un bloc si la boucle n’a pas été interrompue par un ```break```.







````python
i = 0
while i < 5:
    print(i)
    i += 1
else:
    print("Toutes les parcelles analysées avec succès.")
````

###  4.5 Exercice pratique : Simulation d’aménagement urbain

## Objectif
Écrire un programme Python qui simule une analyse d’aménagement urbain pour un nouveau quartier en appliquant les notions suivantes :

- Boucles `for` et `while`
- La fonction `range()`
- Boucles imbriquées
- Itérations sur des listes
- Conditions (`if`, `break`)
- Cumul de données

---

## Contexte
Vous êtes chargé de simuler l’aménagement d’un quartier structuré en **5 zones**, chacune composée de **5 parcelles** de **0.75 km²**. Votre programme doit générer une simulation automatique, avec détection d’une zone à risque et suivi de la congestion urbaine.

---

## Spécifications

1. **Compte à rebours** de 3 secondes avant le lancement.
2. Pour chaque **zone (de 1 à 5)** :
   - Affichage du numéro de la zone.
   - Affichage des **parcelles** et de leur superficie.
   - **Cumul de la superficie totale**.
3. Si la **zone 4** est atteinte, elle est considérée comme **à risque**, et l’analyse s’interrompt.
4. Démarrage d’une **simulation de congestion routière** :
   - Congestion initiale : 0 %
   - Augmentation de **10 % par minute**
   - Arrêt à 80 %


````python
# 1. Compte à rebours avant le début
for i in range(3, 0, -1):
    print(f"Simulation démarre dans {i} secondes")
print("Simulation lancée\n")

# 2. Analyse des zones et des parcelles
superficie_totale = 0
zones = [1, 2, 3, 4, 5]
parcelles_par_zone = 5
superficie_parcelle = 0.75

for zone in zones:
    print(f"Analyse de la zone {zone}")

    # Détection d'une zone à risque
    if zone == 4:
        print(" Zone 4 identifiée comme à risque. Arrêt de l’analyse.")
        break

    for parcelle in range(1, parcelles_par_zone + 1):
        print(f"   Parcelle {parcelle} -> {superficie_parcelle} km²")
        superficie_totale += superficie_parcelle

print(f"\n Superficie totale analysée : {superficie_totale} km²\n")

# 3. Simulation de congestion
congestion = 0
minute = 0

while congestion < 80:
    minute += 1
    congestion += 10
    print(f"Minute {minute} : congestion à {congestion}%")

print("Seuil critique de congestion atteint.\n")
````
